# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

## iris Data Analysis

```python
iris = sns.load_dataset('iris')
```

### sepal length petal length Relationship

```python
sns.relplot(x='petal_length', y='sepal_length', data=iris, kind='line',color='blue')
```

### sepal width petal width Relationship

```python
sns.relplot(x='petal_width', y='sepal_width', data=iris, kind='line',color='blue')
```

### NoteBook Insights

#### from the previous graphas there is a relationship when sepal length increase the petal length increase and no relationship between sepal width and petal width
