# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

## Gammas Data Analysis

```python
gammas = sns.load_dataset('gammas')
```

## timepoint BOLD signal RelationShip

```python
sns.relplot(x='timepoint', y='BOLD signal', data=gammas, kind='line',color='red')
```

## timepoint subject RelationShip

```python
sns.barplot(x='timepoint', y='subject', data=gammas.head(500))
```
